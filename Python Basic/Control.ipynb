```python
#if statement
age=20
if age>=18:
    print("You are Allowed to vote")
```

```python
#else -->It executes a block of code. If the condition in the statement is false
age=6
if age>=18:
    print("You are Allowed to vote")
else:
    print("You are Minor")
```

```python
#elif --> It allows you to check multiple conditions. It stand with else if
age=15
if age<13:
    print("You are Child")
elif age<18:
    print("You are Teenager")
else:
    print("You are Adult")
```

```python
#nested Conditional Statement
num=int(input("Enter the number"))
if num>=0:
    print("Number is Positive")
    if(num %2==0):
        print("This is EVEN")
    else:
        print("This is ODD")
else:
    print("Its Negative number")
```

```python
#usecases
#PRACTICAL EXAMPLE
#if a leap year or not

year=int(input("Enter the Year: "))
if year %4==0:
    if year %100==0:
        if year %400==0:
            print(year,"Its a Leap Year")
        else:
            print(year,"its not a leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

```python
#SIMPLE CALCULATOR USING IF ELSE
num1 =float(input("Enter 1ST Number: "))
num2 =float(input("Enter 2ND Number: "))
operation=input("Enter Operation(+,-,*,/): ")
if operation =='+':
    result =num1+num2
elif operation =='-':
    result =num1-num2
elif operation =='*':
    result =num1*num2
elif operation =='/':
    if num2!=0:
        result =num1/num2
    else:
        result="Error! Division by Zero."
else:
    result ="Invalid Operation"
print(result)

```

```python
#WHILE LOOP
count=0
while count<5:
    print(count)
    count=count+1
```

```python
count=0
while count%2==0:
    print(count)
    count=count+1
```

```python
#LOOP CONTROL STATEMENT
#BREAK STATEMENT ->
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
#continue Statement
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
#Pass statement
for i in range(5):
    if i==3:
        pass
    print(i) 
```

```python
#NESTED LOOP
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
#calculate the sum of first n numbers natural using while and for loop
n=10
sum=0
count=1
while count<=n:
    sum=sum+count
    count=count+1
print("Sum of N natural numbers: ",sum)
```

```python
#DISPLAY PRIME NUMBER BETWEEN 1 AND 100
for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
            else:
                print(num)
```
